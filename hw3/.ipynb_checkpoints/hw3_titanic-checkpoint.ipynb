```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn import svm

train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')
```

```python
#查看資料
train.head()
```

```python
train.info()
```

```python
test.info()
```

# 圖像化資料分布

```python
sns.pointplot(x="Pclass", y="Survived", hue="Sex", data=train,
              palette={"male": "blue", "female": "pink"},
              markers=["*", "o"], linestyles=["-", "--"]);
#票的分類
#可看出存活率和票等級是正相關的
#也明顯看出女性存活率高於男性
```

```python
sns.barplot(x="Embarked", y="Survived", hue="Sex", data=train);
#port of embarkation 從哪裡登船
#從C登船的人存活率似乎較高
```

```python
#選取屬性
seleted_features = ["Pclass","Sex","Age","SibSp","Parch","Fare","Embarked"]

X_train = train[seleted_features]
y_train = train["Survived"]

X_test = test[seleted_features]
```

# 處理缺失值

```python
X_train['Age'].fillna(X_train['Age'].mean(), inplace=True)
X_train['Embarked'].fillna('S', inplace=True)  #X_train['Embarked'].describe() 看出最多是S
X_test['Age'].fillna(X_test['Age'].mean(), inplace=True)
X_test['Fare'].fillna(X_test['Fare'].mean(), inplace=True)
```

```python
X_train.info()
```

```python
X_test.info()
```

# 處理category

```python
X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)
```

```python
svc = svm.SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
svc_survived = svc.predict(X_test)
```

```python
submission = pd.DataFrame({'PassengerId':test['PassengerId'], 'Survived':svc_survived})
```

```python
submission
```

```python
submission.to_csv('submission.csv', index=False)
```
